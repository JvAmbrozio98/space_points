```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("master_combined.csv")
```

```python
df
```

```python
new_df = df.drop(columns=['Unnamed: 0'])
```

```python
new_df
```

```python
new_df['periodo_inicial'] = pd.to_datetime(new_df['periodo_inicial'])
```

```python
new_df['periodo_final'] = pd.to_datetime(new_df['periodo_final'])
```

```python
new_df['month'] = new_df['periodo_inicial'].dt.month
```

```python
new_df['year'] = new_df['periodo_inicial'].dt.year 
```

```python
def apply_season(number):
    if(number) in (12,1,2):
        return 'verao'
    elif(number) in (3,4,5):
        return 'outono'
    elif(number) in (6,7,8):
        return 'inverno'
    elif (number) in (9,10,11):
        return 'primavera'
```

```python
new_df['season'] = new_df['month'].apply(apply_season)
```

```python
new_df
```

```python
new_df = new_df.groupby(["cidade", "latitude","longitude", "season"]).agg({
    "ndvi_medio": "mean",
    "evi_medio": "mean",
    "leaf_area_index_medio": "mean",
    "sm_rootzone_pctl_medio": "mean",
    "sm_profile_pctl_medio": "mean",
    "lst_day_1km_c": "mean"
}).reset_index()
```
